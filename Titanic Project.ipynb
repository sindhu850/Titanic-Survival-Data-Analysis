```python
def male_female_child(passenger):
    age,sex=passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
##create new colunm to apply the above code
titanic_df['person']=titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
deck=titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels=[]
for level in deck:
    levels.append(level[0])

cabin_df=DataFrame(levels)

cabin_df.columns=['Cabin']

sns.catplot(x='Cabin',hue='Cabin',data=cabin_df,palette='winter_d',kind='count',legend=False)
```

```python
cabin_df=cabin_df[cabin_df.Cabin != 'T']

sns.catplot(x='Cabin',hue='Cabin',data=cabin_df,palette='summer',kind='count',order=['A','B','C','D','E','F','G'],legend=False)
```

```python
## who was alone and who is with family ?
titanic_df.head()
```

```python
titanic_df['Fstatus']=titanic_df.SibSp+titanic_df.Parch
```

```python
titanic_df['Fstatus']
```

```python
titanic_df['Fstatus']=titanic_df['Fstatus'].astype('object')
titanic_df['Fstatus'].loc[titanic_df['Fstatus'] != 0 ] = 'with family'
titanic_df['Fstatus'].loc[titanic_df['Fstatus']==0] = 'alone'
```

```python
titanic_df.tail()
```

```python
sns.catplot(x='Fstatus',hue='Fstatus',data=titanic_df,palette='Blues',kind='count',legend=False)
```

```python
titanic_df['Survivor']=titanic_df.Survived.map({0:'no',1:'yes'})

sns.catplot(x='Survivor',hue='Survivor',data=titanic_df,palette='Set1',kind='count')
```
